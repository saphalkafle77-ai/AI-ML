```python
import requests

```

```python
#Requests
URL = "https://www.scrapethissite.com/pages/simple/"
res = requests.get(URL)
```

```python
if res.status_code == 200:
    # print(res.text) #html
    # print(res.content)  #raw data
    print(res.headers)  #metadata,servers

with open("scraped_data/data1.html","w",encoding="utf-8") as f:
    f.write(res.text)
```

```python
#beautiful soup
from bs4 import BeautifulSoup
with open("scraped_data/data1.html","r") as f:
    html_content = f.read()

soup = BeautifulSoup(html_content, "lxml")
```

```python
soup.find("h3")  #1st matching element
all_h3=soup.find_all("h3")

all_countries = []
for h3 in all_h3:
    country_name = h3.get_text(strip = True)
    # print(h3.find_parent("div") ["class"])

    population = (h3.find_next("div").select("span.country-population") [0].get_text(strip = True)) #from select
    # print(h3.attrs)  use attribute

    all_countries.append([country_name,population])
```

```python
all_countries
```

```python
import pandas as pd
df = pd.DataFrame(all_countries,columns = ["Name","Population"])
```

```python
#after data are clean
df.to_csv("cleaned_data/data.csv",index = False)
```
