```python
import json
```

```python
#load the data
def load_data(filename):
    with open(filename,"r") as f:
        data = json.load(f)
    return data
```

```python
data = load_data("store_data.json")
print(data,type(data))
```

```python
#data is inconsistent - mixedtypye(rating)
#missing data (age)
#duplicated data
```

```python
#clean and structure of data
def clean_data(data):
    text_to_num = {"one":1,"two":2,"three":3,"four":4,"five":5}
    cleaned_data = []
    unique_users = set()
    for user in data:
        #clean rating
        raw_rating = str(user["rating"]).strip().lower()
        
        if raw_rating in text_to_num:
            raw_rating = text_to_num[raw_rating]
        user["rating"] = raw_rating

       #handle missing value
        
        raw_age = user.get("age")
        if(raw_age == None):
            user["age"] = None

        #deduplication
        if(user["name"].strip() in unique_users):
            continue
        unique_users.add(user["name"])
        
        cleaned_data.append(user)
    return cleaned_data
```

```python
data = clean_data(data)
```

```python
#get meaningful insight
def get_insight(data):

    #avg rating
    total_rating = 0
    for user in data:
        total_rating += float(user["rating"])
    average = total_rating/ (len(data))
    print(average)

    #percentage of user with poor rating
    poor_user = 0
    for user in data:
        if (float(user["rating"]) < 3):
            poor_user +=1
            print(poor_user/len(data)*100)
            
```

```python
get_insight(data)
```

```python
#Build  the recommendation feature
def get_recommendation(data):
    recommendation = []
    for user in data:
        curr_recomm = {}
        curr_recomm["name"] = user["name"]
        if(float(user["rating"]) >= 4):
            curr_recomm["brand"] = "Apple"
          
        else:
            curr_recomm["brand"] = "samsung"
      recommendation.append(curr_recomm)   
    return recommendation
```

```python
get_recommendation(data)
```
