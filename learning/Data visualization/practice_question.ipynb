```python
import matplotlib.pyplot as plt
```

```python
days = ["Mon", "Tue", "Wed", "Thu", "Fri"]

cities = ["New York", "London", "Delhi", "Tokyo"]

temperatures = [
    [22, 23, 21, 24, 25],   # New York
    [18, 19, 17, 20, 21],   # London
    [30, 32, 31, 33, 34],   # Delhi
    [25, 26, 24, 27, 28]    # Tokyo
]

fig,ax = plt.subplots(2,2)

# city_no = 0
# for i in range(2): #rows (i) -0,1
#     for j in range(2):#columns
#         ax[i][j].plot(days,temperatures[city_no],marker = "o")
#         ax[i][j].set_title(cities[city_no])
#         ax[i][j].grid(True)
#         city_no = city_no + 1



#alternative
axes = ax.flatten()
for i,axe in enumerate(axes):
    axe.plot(days,temperatures[i],marker = 'o')
    axe.set_title(cities[i])
    axe.grid(True)
fig.suptitle("Temperature in Cities over the week")
fig.supylabel("Temperature")
fig.supxlabel("Days")
fig.tight_layout()
```
