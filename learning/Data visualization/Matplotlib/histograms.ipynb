```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(0)
score = np.random.normal(70,10,100)  #normal distribution (mean,sd)
bins = [0,20,70,90,100]
plt.hist(score,color = "skyblue",edgecolor = "black",bins = bins)
plt.ylabel("# of students")
plt.xlabel("scores")
```

```python
marks = [
     25, 38, 42, 55, 61, 67, 72, 78, 83, 88,
    91, 95, 47, 53, 59, 64, 69, 74, 79, 85,
    92, 97, 33, 45, 58, 62, 71, 76, 81, 89
]
plt.hist(marks,color = "aqua",edgecolor = "black",bins = bins,linewidth = "2")
#axvline
plt.axvline(x=33,label = "passing marks",color = "red")
plt.xlabel("Marks")
plt.ylabel("no of students")
plt.title("distribution of marks for students")

```

```python
#multiple dataset on single histogram
legit_transactions = [
    2.99, 5.49, 8.99, 12.50, 14.99, 19.99, 23.45, 29.99, 34.99, 39.50,
    45.00, 49.99, 55.25, 60.00, 75.99, 89.99, 120.50, 150.00, 199.99,
    249.99, 300.75, 450.00, 600.00, 850.00, 1200.00
]

fraud_transactions = [
    50, 100, 150, 200, 300, 500, 500, 750, 1000, 1000,
    1200, 1500, 1500, 2000, 2500, 3000, 3000
]


plt.hist(fraud_transactions,color = "red",edgecolor = "black",label = "fraud",bins = 20,alpha = 0.5)
plt.hist(legit_transactions,color = "aqua",edgecolor = "black",label = "legit",bins = 20,alpha = 0.5)
plt.title("fraud vs legit transaction")
plt.xlabel("no of transaction")
plt.ylabel("Amount in $")
plt.grid(True)
plt.legend()
```
