```python
#real data

#oscar winner movies
oscar_movies = [
    "The dark knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Agro"
]
years = [2008,2009,2010,2011,2012]
oscar_revenue = [1005,170,437,133,232]#in$M

#plot
plt.plot(years,oscar_revenue)
plt.title("Oscar movies revenue in each year")
plt.xlabel("years")
plt.ylabel("revenue in $M")
```

```python
#Non oscar winner movies
non_oscar_movies = ["slumdog Millionare","Avatar","Inception","Hugo","Lincoln"]
non_oscar_revenue = [378,2788,829,186,275] #in $M\

plt.plot(years,oscar_revenue,label="oscar winners")
plt.plot(years,non_oscar_revenue,label = "Non-oscar winners")
plt.title("oscar movies revenue vs Non-oscar movies Revenue")
plt.xlabel("years")
plt.ylabel("Revenue (in $M) ")

plt.legend()
```

```python
#format string
non_oscar_movies = ["slumdog Millionare","Avatar","Inception","Hugo","Lincoln"]
non_oscar_revenue = [378,2788,829,186,275] #in $M\

plt.plot(years,oscar_revenue,"o--r",label="oscar winners")
plt.plot(years,non_oscar_revenue,color="black",marker="<",linestyle = "--",label = "Non-oscar winners")
plt.title("oscar movies revenue vs Non-oscar movies Revenue")
plt.xlabel("years")
plt.ylabel("Revenue (in $M) ")

plt.legend()
```
