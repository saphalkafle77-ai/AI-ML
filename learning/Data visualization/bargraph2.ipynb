```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#comparing of two data set
#oscar winner movies

oscar_movies = [
    "The dark knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Agro"
]

oscar_revenue = [1005,170,437,133,232]#in$M

non_oscar_movies = ["slumdog Millionare","Avatar","Inception","Hugo","Lincoln"]
non_oscar_revenue = [378,2788,829,186,275] #in $M\

years = [2008,2009,2010,2011,2012]
x = np.arange(len(years))
width = 0.4

plt.bar(x-width/2,oscar_revenue,width,label="oscar winners")
plt.bar(x+width/2,non_oscar_revenue,width,label= "non oscar winners")
plt.title("oscar vs non-oscar movies revenue")
plt.xlabel("years")
plt.ylabel("revenue")
plt.legend()
plt.xticks(x,years)
```
