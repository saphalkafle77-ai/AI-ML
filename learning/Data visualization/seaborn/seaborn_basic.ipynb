```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
sns.set_theme() #style = "white"
tips = sns.load_dataset("tips")
#relational plot
sns.relplot(
    data = tips,
    x="total_bill",
    y="tip",
    hue="smoker",
    size = "size",
    style = "smoker"
)
```

```python
x_vals = [1,2,3,4,5,6,7,8,9]
y_sq = [i**2 for i in x_vals]
sns.set_theme(style = "ticks")
sns.relplot(
    
    x= x_vals,
    y= y_sq,
    
    kind = "line"
    
)
```
