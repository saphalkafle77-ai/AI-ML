```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_theme() #style = "white"
#scatter
tips = sns.load_dataset("tips")
sns.scatterplot(
    data = tips,
    x="total_bill",
    y="tip",
    hue="smoker",
    size = "size",
    style = "smoker"
)
```

```python
#lineplot
flights = sns.load_dataset("flights")
sns.lineplot(
    data = flights,
    x = "year",
    y = "passengers"
)
```
