```python
#plot y = √x
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 100, 100_000)
y = np.sqrt(x)

# Plot
plt.plot(x, y, label="y = √x")
plt.xlabel("x")
plt.ylabel("y")
plt.title("y = √x")
plt.legend()
plt.grid(True)

plt.show()
```

```python
# Create x values from -2π to 2π
x = np.linspace(-2 * np.pi, 2 * np.pi, 100)
y = np.sin(x)

# Plot the graph
plt.plot(x, y, label='y = sin(x)', color='blue')

plt.axvline(0, color='black', linewidth=1.5, linestyle='--', label='x = 0')
plt.axhline(0, color='black', linewidth=1.5, linestyle='--', label='y = 0')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of y = sin(x)')
plt.grid(True)
plt.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)
y = np.sin(x) + 2

fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(x, y, color='royalblue', linewidth=2, label=r'$f(x) = \sin(x) + 2$')

# Reference lines
ax.axhline(y=2, color='gray', linestyle='--', linewidth=1, alpha=0.6, label='Midline (y = 2)')
ax.axhline(y=3, color='salmon', linestyle=':', linewidth=1, alpha=0.6, label='Max (y = 3)')
ax.axhline(y=1, color='green', linestyle=':', linewidth=1, alpha=0.6, label='Min (y = 1)')
ax.axvline(x=0, color='black', linewidth=0.8, alpha=0.5)

# Formatting
ax.set_title(r'Graph of $f(x) = \sin(x) + 2$', fontsize=16, fontweight='bold', pad=15)
ax.set_xlabel('x (radians)', fontsize=13)
ax.set_ylabel('f(x)', fontsize=13)
ax.set_xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi])
ax.set_xticklabels([r'$-2\pi$', r'$-\pi$', '0', r'$\pi$', r'$2\pi$'], fontsize=12)
ax.set_ylim(-0.5, 4.5)
ax.legend(fontsize=11)
ax.grid(True, linestyle='--', alpha=0.4)

plt.tight_layout()

plt.show()
```
