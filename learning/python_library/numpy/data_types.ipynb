```python
import numpy as np
```

```python
#data types of numpy
#integer
arr = np.array ([1,2,3,4])
print(arr,arr.dtype)
```

```python
#float
arr = np.array ([1.3,2,3,4])
print(arr,arr.dtype)
```

```python
#string
arr = np.array ([1.3,2,3,4,"safal"])
print(arr,arr.dtype)
```

```python
#complex numbers
arr = np.array([3 + 5j])
print(arr,arr.dtype)
```
