```python
import numpy as np
```

```python
#reshaping
arr = np.array([[1,2,3] , [4,5,6]])
print(arr)
reshape = arr.reshape((3,2))
print(reshape,reshape.shape)

#flatten
flattened = arr.flatten()
print(flattened)
```

```python
#indexing in array
arr = np.array([1,2,3,4,5])
print(arr[0])
arr2 = np.array([[1,2,3],[4,5,6]])
print(arr2[1][1])
```

```python
#fancy indexing
arr = np.array([1,2,3,4,5])
idx = [1,2,4]
print(arr[idx])
```

```python
#boolean indexing
arr = np.array([1,2,3,4,5])
print(arr[arr >3])
```

```python
#slicing
arr = np.array([1,2,3,4,5])
print(arr[1:4])
```

```python
#copy Vs view
#python list
# num = [1,2,3,4,5]
# sub_list = num[1:3]
# print(sub_list)
# sub_list[0] = 200
# print(sub_list)
# print(num)

#numpy array

arr = np.array([1,2,3,4,5])
sub_arr = arr[1:3].copy()
print(sub_arr)
sub_arr[0] = 200
print(sub_arr)
print(arr)

```
