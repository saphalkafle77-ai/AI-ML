```python
import numpy as np
```

```python
#mathematical fnx

#aggregation
arr = np.array([1,2,3,4,5])
print(np.sum(arr))
print(np.prod(arr))
print(np.min(arr))
print(np.argmin(arr))
```

```python
#power function
print(np.square(arr))
print(np.sqrt(arr))
print(np.pow(arr,3))
```

```python
#log and exponential
print(np.log(arr))
print(np.exp(arr))
```

```python
#rounding
print(np.round(3.14))
print(np.ceil(3.14))
print(np.floor(3.14))
print(np.trunc(3.14))
print(np.abs(arr))  #absoulte
```
