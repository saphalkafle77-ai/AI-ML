```python
import pandas as pd
```

```python
#dataframe by dictionary
info = {
     "name": ["safal","ram","shyam"],
     "age" : [21,22,23],
     "cgpa": [9.2,8.5,8.9]
 }
df = pd.DataFrame(info)
print(df)


#print column
print(df.columns)

#for rows label
print(df.index)
```

```python
#dataframe from list of list
df = pd.DataFrame([["safal",23],["rita",21],["sita",31]],columns = ["Name","age"])
print(df)
```

```python
#Dataframe from Numpy array
import numpy as np
np_array = np.array([[1,2,3],[4,5,6]])
df = pd.DataFrame(np_array,columns = ["a","b","c"])
print(df)
```
