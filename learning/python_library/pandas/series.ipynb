```python
import pandas as pd
```

```python
s = pd.Series([1,2,3,4,5])
print(s)
print(type(s))


#accessing value
#indexing
print(s[3])

```

```python
#label
s = pd.Series([11,12,13,14,15],index= ["Adam","saphal","hari","sita","rita"])
print(s)
print(s["Adam"])
```

```python
#vectorized operation
s1 = pd.Series([1,2,3,4])
s2 = pd.Series([10,20,30,40])
print(s1 + s2)
s1[0] = 100
print(s1)
```

```python
#mutable value , immutable size
s1[0] = 100
print(s1)

changed_s1 = s1.drop(0)
print(s1)
print(changed_s1)
```
