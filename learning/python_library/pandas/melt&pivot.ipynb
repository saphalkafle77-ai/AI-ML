```python
import pandas as pd
```

```python
df=pd.DataFrame({"country":["USA","USA","India","India"],
                 "year":[2020,2021,2020,2021],
                 "sales":[100,120,90,110],
                 "profit":[20, 25, 18, 22],
                 
                 
                })
melted_data = df.melt(
    id_vars=["country","year"],
    value_vars=["sales","profit",],
    var_name="metrics",
    value_name="value"
)

original = melted_data.pivot(
    index=["country","year"],
    columns="metrics",
    values="value"
)
original
```
