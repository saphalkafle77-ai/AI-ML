```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
fig,ax = plt.subplots()
sns.lineplot(
    data = tips,
    x="day",
    y="total_bill",
    hue = "sex",
    marker = "o",
    # errorbar = None,
    ax =ax
)
ax.set_title("avg bill value for each day")
ax.set_xlabel("Days")
ax.set_ylabel("Total Bill")
fig.tight_layout()
```
