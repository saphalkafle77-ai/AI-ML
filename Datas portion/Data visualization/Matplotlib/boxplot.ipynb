```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = [7,8,5,6,9,4,10,12,15]
plt.boxplot(data)
plt.grid(True)
```

```python
#multiple datasets
group1 = np.random.normal(50,10,100)
group2 = np.random.normal(60,10,100)
plt.boxplot([group1,group2],tick_labels =["group1","group2"] ,showmeans = True)
plt.grid(True)
```
