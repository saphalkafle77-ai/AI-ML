```python
import matplotlib.pyplot as plt
#barchart
countries = [
    "Nepal",
    "India",
    "china",
    "usa"
]
population = [100,405,520,203]
year = [2008,2009,2010,2011]

plt.bar(countries,population,color = "lightblue")
plt.title("population growth in each year")
plt.xlabel("countries")
plt.ylabel("population growth")

for i in range(len(countries)):
    plt.text(countries[i],population[i]+10,str(population[i]),ha="center")
plt.ylim(0,max(population) + 100)
```
