```python
import matplotlib.pyplot as plt
```

```python
#piecharts
expenses = ["salaries","rent","marketing","R&D","Miscellaneous"]
amounts = [500,150,120, 100, 50]
plt.style.use("ggplot")


#customization
explode = [0,0,0.4,0,0]
plt.pie(amounts,labels = expenses,autopct = "%1.1f%%",
       wedgeprops = {
           "edgecolor":"black",
           "linewidth":1,
           
       }, explode = explode,shadow = True)
```
