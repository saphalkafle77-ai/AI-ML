```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import 
```

```python
#1
score = np.random.normal(loc=70, scale = 10,size = 200)
mean_score = np.mean(score)
plt.hist(score,edgecolor = "black",bins =20)

plt.axvline(x= mean_score,label = "passing score",color = "red")
plt.title("Histogram of scores")
plt.xlabel("score")
plt.ylabel("frequency")
plt.legend()
```

```python
#2
penguins = sns.load_dataset("penguins")
sns.scatterplot(
    data = penguins,
    x = "flipper_length_mm",
    y = "body_mass_g",
    hue = "species"
)
plt.title("Flipper_lengthvs bodymass")
```

```python
#3
tips = sns.load_dataset("tips")

sns.boxplot(
    data = tips,
    x = "day",
    y = "total_bill",
    hue = "sex"
)
median = tips.groupby("day")["total_bill"].median().idxmax()
median
#outlier are the point outside of our value range determining the value is low or high
```

```python
#4
sns.scatterplot(
    data = tips,
    x = "total_bill",
    y = "tip",
    hue = "sex",
    style = "smoker"
)
plt.title("Total_bill vs Tips")
plt.xlabel("tip")
plt.ylabel("Total_bill")
tips.groupby("smoker")["tip"].mean()
```

```python
#q5
avg_tip_pivot = tips.pivot(index = "day",columns = "time",values = "tip",aggfunc = "mean")
sns.heatmap(
    data = avg_tip_pivot,
    cmap = "coolwarm",
    annot = True,
    fmt = "d"
)
plt.title("average tip")
plt.tight_layout()
```
