```python
import gymnasium as gym
import numpy as np
import random
```

```python
# Sample env
env = gym.make("CliffWalking-v1")
```

```python
print(env.observation_space.n) #states
print(env.action_space.n) #actions

starting_state, _ = env.reset()  #give initial state 36 = 3*12+0
print(starting_state)
```

# SARSA

```python
gamma = 0.99
alpha = 0.5
epsilon = 0.1
episodes = 500 
```

```python
# Q-table = action value
Q = np.zeros((48,4))
```

```python
# policy - Epsilon-greedy : on the basis of state it says action

def epsilon_greedy(state):
    if random.random() < epsilon:
        return env.action_space.sample() #random action ==> Explore
    else:
        return np.argmax(Q[state]) #exploit
```

```python
for episode in range(episodes):
    render = (episode % 50 ==0)

    if render:
        env = gym.make("CliffWalking-v1", render_mode = "human")
        print(f"rendering env for episode = {episode + 1}/500 ")
    else:
        env = gym.make("CliffWalking-v1")
        
    done = False # when the episode has ended
    state,_ = env.reset()
    action = epsilon_greedy(state)

    total_reward = 0
    episode_len = 0 #how many moves

    while not done:
        next_state,reward,terminated,truncated, _ = env.step(action)
        done = terminated or truncated
        next_action = epsilon_greedy(next_state)

        #SARSA update
        Q[state,action] += alpha *(reward + gamma*(Q[next_state,next_action]- Q[state,action]))

        state = next_state #for next step
        action = next_action

        total_reward += reward
        episode_len +=1
    print(f"episode = {episode + 1}/500 : totalreward = {total_reward} & ep_length = {episode_len}")
    env.close()
```

# What did our agent learn?

```python
env = gym.make("CliffWalking-v1",render_mode = "human")
state,_ = env.reset()
done = False
total_reward = 0
episode_len = 0

while not done:
    action = np.argmax(Q[state])
    next_state , reward, terminated, truncated, _ = env.step(action)
    done = terminated or truncated
    state = next_state


    total_reward += reward
    episode_len +=1

print(f"total reward = {total_reward} & episode len = {episode_len}")
env.close()
```
