```python
import numpy as np
```

```python
#linear regression
class LinearRegression:
    def __init__(self,learning_rate = 0.01, n_iter = 1000):
        self.bias = None
        self.coefficient = None
        self.lr = learning_rate
        self.n_iter = n_iter

    def fit(self,X,y):
        m,n = X.shape #no of sample and no of features
        #step 1
        self.bias = 0
        self.coefficient = np.zeros(n)  #return arrays of zeros

        for i in range(self.n_iter):

            #step 2
            y_pred = self.bias + np.dot(X,self.coefficient)
    
            #step 3 (calculating gradient)
            db = (1/m) * np.sum(y_pred - y)
            dc = (1/m) * np.dot(X.T,(y_pred - y))
    
            #step 4 - convergence theorem parameter update
            self.bias = self.bias - self.lr * db
            self.coefficient = self.coefficient - self.lr * dc


    def predict(self,X):
        y_pred = self.bias + np.dot(X,self.coefficient)
        return y_pred 
```

```python
X = np.array([[1],[2],[3],[4],[5]])
y = np.array([2,4,6,8,10])

model = LinearRegression()
model.fit(X,y)
y_pred = model.predict(X)
print(y_pred)
print(model.bias)
print(model.coefficient)
```
